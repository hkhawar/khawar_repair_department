## Directories and Filepath

```python
def paths(topdir, subdir, figures):
    
    folderpath = os.path.join(Path("../", topdir), subdir)

    figpath = os.path.join(folderpath, figures)


    def create_folder(path):
             
        if not os.path.exists(path):
        
            print ("directory is created")
        
            return os.makedirs(path)
    
        else:
            
            return print ("directory is already existed")

    
    
    folder = create_folder(folderpath)

    figdir = create_folder(figpath)
    
    return folderpath, figpath


folderpath, figpath = paths(topdir, subdir, figures)
print(folderpath)
print(figpath)
```

## Defining Paths

```python
def csv_paths(folderpath):
    
    csvpath = [os.path.join(folderpath, f) for f in os.listdir(folderpath)]
    
    apriso_path = [f for f in csvpath if "Apriso" in f][0]
    
    tracker_path = [f for f in csvpath if "Tracker" in f][0]
    
    buffer_path = [f for f in csvpath if "Buffer" in f][0]
    
    adjkpi_path = [f for f in csvpath if "Adjust" in f][0]
    
    return apriso_path, tracker_path, buffer_path, adjkpi_path


apriso_path, tracker_path, buffer_path,adjkpi_path = csv_paths(folderpath)

print(apriso_path)
print(tracker_path)
print(buffer_path)
print(adjkpi_path)
```

### Creating Merged dataframe which resulted after matching Tracker and Apriso Data

```python
def create_merged_csv(apriso_path, tracker_path, figpath):
    
    tracker_columns = ['DATE','RBU_SN', 'FAIL_Category', 'Fail_Group', 
                       'TRSH_NAME','STATUS', 'Notes', 'REPAIR_Person',
                       'REPAIR_NOTES', 'PRE-TRSH_Person']
    dt_tracker = (pd.read_csv(tracker_path, engine='python')
                        .loc[:, tracker_columns]
                       )

    dt_tracker = dt_tracker[dt_tracker['RBU_SN'].notnull()]

    #dt_tracker.dropna(subset=['RBU_SN'], inplace=True)

    dt_tracker['DATE'] = dt_tracker['DATE'].fillna(method='ffill')

    dt_tracker['DATE'] = pd.to_datetime(dt_tracker['DATE'], dayfirst = True,  errors='coerce')

    data_tracker = dt_tracker.loc[(dt_tracker['DATE'] >= startdate) & (dt_tracker['DATE'] <= enddate)]    
    
    pattren1 = '|'.join(["ea2a", " ea2a" , " EA2A"])
    
    pattren2 = '|'.join(["e23", " e23" , " E23"])
    
    data_tracker['RBU_SN'] = data_tracker['RBU_SN'].str.replace(pattren1, "EA2A")
    
    data_tracker['RBU_SN'] = data_tracker['RBU_SN'].str.replace(pattren2, "E23") 
    
    
    apriso_columns = ['Serial #', 'Level 1', 'Level 2', 'Level 3', 'ProductNo']
    
    data_apriso = (pd.read_csv(apriso_path)
                   .loc[:, apriso_columns]
                  )


    data_apriso['RBU_SN'] = data_apriso[data_apriso.columns[0:4]].apply(
        lambda x: ''.join(x.dropna().astype(str)),
        axis=1
    )
    
    data_apriso = (data_apriso.loc[:, ['RBU_SN', 'ProductNo']])
    
    #data_tracker[data_tracker['DATE'].str.contains("\n\n")].replace("\n\n", " ")
    

    merged = data_tracker.merge(data_apriso, how='left', left_on='RBU_SN', right_on= 'RBU_SN')

    merged = (merged
              .drop_duplicates()
              .drop(columns=['Fail_Group', 'Notes'], axis=1)
             )
              

        
    merged.loc[(merged['ProductNo'].str.contains("ROA1286713")) & (merged['FAIL_Category'] == 'RF'), 'FAIL_Category'] = "MBFT"
    

    
    
        

    
    return merged

merged = create_merged_csv(apriso_path, tracker_path, figpath)

merged.head()
```

## TRSH Performance and KPI

```python
def TRSH_Performance_KPI(merged, TRSH_NAME, figpath, DAYS, adjusted_kpi):
    
    
    """ 
    Here we are removing all cases where STATUS is equal to prescrap as this information should not be
    included in the calculation of KPI
    
    """
    tmp = merged.copy()
    
    
    def prescrap(x):

        if (x['STATUS'] == 'Pre-Scrap'):
            return np.nan

        else: 
            return x["FAIL_Category"]


    tmp['RemovePrescrap']   =  tmp.apply(prescrap, axis=1)  


    cat_condition=["DIG", "RF", "MBFT", "ICO_TRSH", "VRT", "GU", "Burn-In_Test", "External_Claim", 'Re-work_Order']
    
     
        # Generating dataframe for repair department report
   
    TRSH_report_data = (tmp.query('TRSH_NAME in @ TRSH_NAME and RemovePrescrap in @ cat_condition')
                 .dropna(subset=['ProductNo']) ## droping na values based on ProductNo column
                 .drop_duplicates(subset=["RBU_SN", "TRSH_NAME","RemovePrescrap", "ProductNo"], keep='last')  ## drop duplicate units and keeping the last row with unique unit
                  )
    
    # Calculating unique number of attempts

    tmp_df = (tmp.query('TRSH_NAME in @ TRSH_NAME and RemovePrescrap in @ cat_condition')
                 .dropna(subset=['ProductNo']) ## droping na values based on ProductNo column
                 .drop_duplicates(subset=["RBU_SN", "TRSH_NAME","RemovePrescrap", "ProductNo"], keep='last')  ## drop duplicate units and keeping the last row with unique unit
                 .groupby(['TRSH_NAME',"RemovePrescrap"])
                 .count()
                 .reset_index()
                 .iloc[:, 0:3]  # DATE is counting the number of product against person
                 .rename(columns={'TRSH_NAME': 'Names',
                                 'RemovePrescrap': "Category", 
                                 'DATE': "Count",
                                }
                        )
                )


    # Removing Prescrap counting in total attempts again respective TRSH name
    tmp1 = tmp[tmp["RemovePrescrap"].notna()]

   # Calculating total number of attempts

    total = (tmp1.query('RemovePrescrap in @ cat_condition')
                .groupby('TRSH_NAME').count() 
                .reset_index()
                .loc[:, ['TRSH_NAME','DATE']]
                .rename(columns={'TRSH_NAME': 'Names',
                                 'DATE': "Count",
                                })
                )



    total['Category'] = 'Total_Attempts'

    total = total.loc[:, ['Names', 'Category', 'Count']]

    fn = (tmp_df.append(total)
          .reset_index()
          .drop(columns='index')
         )






    p = gg.ggplot(fn, gg.aes(x='Names', y='Count',fill="Category", label='Count')) + \
        gg.geom_col(stat="identity") + \
        gg.theme_bw() + \
        gg.coord_flip() + \
        gg.geom_text(gg.aes(y = "Count",label = "Count"),  
                 size=8, nudge_x=0, nudge_y=10, show_legend=True) + \
        gg.labs(title= 'Troubleshooters Report' + ' (' + str(month) + ')  \n\n No of Days: ' + str(DAYS))



    p = p + gg.facet_wrap('~ Category', nrow=2)



    gg.ggsave(filename='Trouble_shooters_performance_' + str(month) +'.png', plot = p, path = figpath)


    # Calculating TRSH KPI

#     min_threshold_radio=25
#    target_radio= 6 * DAYS    ## 70 min / radio

#     min_threshold_board=25
#     target_board= 8 * DAYS       ## 50 min / board



    def repair_dept_kpi(x):

        if (x['Category'] == 'RF') | (x['Category'] == 'DIG') | (x['Category'] == 'GU') | (x['Category'] == 'Burn-In_Test') | (x['Category'] == 'External_Claim') | (x['Category'] == 'Re-work_Order') | (x['Category'] == 'VRT'):
            return "Radio"

        elif (x['Category'] == 'MBFT') | (x['Category'] == "ICO_TRSH"):
            return "Board" 



   

    kpi = tmp_df.copy()


    kpi = kpi.merge(adjusted_kpi, on='Names')


    def KPI_radio(x):

        if (x['Category'] == "Radio"):

            target_radio= 6 * x['WD_'+month]   ## Target for Radios

            k = (x['Count']) / (target_radio)

            return k 



    def KPI_board(x):

        if (x['Category'] == "Board"):

            target_board= 8 * x['WD_'+month]   ## Target for Board

            k = (x['Count']) / (target_board)

            return k





    kpi['Category'] = kpi.apply(repair_dept_kpi, axis=1)



    kpi['KPI_Radio'] = kpi.apply(KPI_radio, axis=1)



    kpi['KPI_board'] = kpi.apply(KPI_board, axis=1)





    kpi = (kpi.groupby("Names")
               .sum()
               .reset_index()
               .assign(Total_Kpi = lambda x: np.round(x['KPI_Radio'] + x['KPI_board'], 2))
              )

    
    average = np.round(kpi['Total_Kpi'].sum()/ len(TRSH_NAME),2)

    g = gg.ggplot(kpi, gg.aes(x='Names', y='Total_Kpi',fill="Names", label='Total_Kpi')) + \
        gg.geom_col(stat="identity") + \
        gg.theme_classic() + \
        gg.geom_text(position=gg.position_stack(vjust=0.5), size = 8) + \
        gg.geom_hline(yintercept = average,
                          color='#b15928',    
                          size=1,            
                          linetype="dashed") + \
        gg.annotate("text", label = "Avg: " + str(average), color="#de2d26", size = 10, x = 1.4, y =0.7) + \
        gg.labs(title= 'Troubleshooters KPI Report'+ ' (' + str(month) + ')  \n\n No of Days: ' + str(DAYS),
                #x=str(month),
                x=" ",
                y="KPI Index"
               ) + \
        gg.theme(axis_text_x=gg.element_blank())


    gg.ggsave(filename='Trouble_shooters_KPI_' + str(month) + '.png', plot = g, path = figpath)


    return TRSH_report_data, tmp_df, p, g



TRSH_report_data, TRSH_data, TRSH_fig_1, TRSH_fig_2 = TRSH_Performance_KPI(merged, TRSH_NAME,figpath, DAYS, adjusted_kpi)

print(TRSH_fig_1)
print(TRSH_fig_2)
      
```

### Pre-TRSH performance and KPI

```python
def pre_TRSH_Performance_KPI(merged, figpath, PRETRSH_NAME, DAYS, adjusted_kpi):
    
    
    
    def repair_dept_pre_TRSH(x):


        if (x['FAIL_Category'] == "Mech") & (x['STATUS'] == "ALT2"):
            return "ALT2"

        elif (x['FAIL_Category'] == "Mech") & (x['STATUS'] != "ALT2"):
            return "Mech"
        
        elif (x['STATUS'] == 'Pre-Scrap'):
            return np.nan
        
        else:
            return x['FAIL_Category']




        
# ## Calling RF and DIG. as Radio and count twice for similar person   
    

    tmp = merged.copy()


    tmp['Category'] = tmp.apply(repair_dept_pre_TRSH, axis=1)
    

    cat_condition=["Prod_Disass","DIG", "RF", "VRT", "GU", "Burn-In_Test", "External_Claim", 'Re-work_Order', 'ALT2', 'Mech']



    tmp = (tmp[tmp['Category'].isin(cat_condition) & tmp['PRE-TRSH_Person'].isin(PRETRSH_NAME)]
          
           
          )
    
    
    
    
    
    """IN PRE-TRSH case we are removing duplicate Serial Numbers"""
    
    
    tmp1 = (tmp.dropna(subset=['ProductNo']) ## droping na values based on ProductNo column
               .drop_duplicates(['RBU_SN', 'PRE-TRSH_Person', 'Category', 'ProductNo'], keep='last')
           )
    
    preTRSH_report_data = tmp1.copy()




    
    pre_TRSH = (tmp1.groupby(['PRE-TRSH_Person','Category'])
            .count()
            .reset_index()
            .iloc[:, 0:3]
            .rename(columns={'PRE-TRSH_Person': 'Names',
                             'FAIL_Category':"Category",
                             'DATE': "Count"})

                  )




    pre = gg.ggplot(pre_TRSH, gg.aes(x='Names', y='Count',fill="Category", label='Count')) + \
            gg.geom_col(stat="identity") + \
            gg.theme_bw() + \
            gg.coord_flip() + \
            gg.geom_text(gg.aes(y = "Count",label = "Count"),  
                     size=8, nudge_x=0, nudge_y=5, show_legend=True) + \
            gg.labs(title= 'Pre_Troubleshooters Report' + ' (' + str(month) + ')  \n\n No of Days: ' + str(DAYS))



    pre = pre + gg.facet_wrap('~ Category', nrow=2)


    gg.ggsave(filename='Pre_Trouble_shooters_performance_' + str(subdir) +'.png', plot = pre, path = figpath)


    PreTRSH_KPI = (tmp1.groupby(['PRE-TRSH_Person','Category', 'ProductNo'])
            .count()
            .reset_index()
            .iloc[:, 0:4]
            .rename(columns={'PRE-TRSH_Person': 'Names',
                             'FAIL_Category':"Category",
                             'DATE': "Count"})

                  )
    




    def KPI_preTRSH(x):
         
        if x['ProductNo'].startswith("KRC"):    # Future product in case if timings is different
            
            target_Radio = 15 * x['WD_'+month]

            k = (x['Count']) / target_Radio

            return k

        elif (x['Category'] == "Prod_Disass") | (x['Category'] == "RF") | (x['Category'] == "DIG") | (x['Category'] == "VRT") | (x['Category'] == "GU") | (x['Category'] == 'Burn-In_Test') | (x['Category'] == 'Mech') | (x['Category'] == "External_Claim"):
            
            target_Radio = 15 * x['WD_'+month]

            k = (x['Count']) / target_Radio

            return k

        elif (x['Category'] == 'ALT2'):
            
            target_ALT2 = 28 * x['WD_'+month]

            k = (x['Count']) / target_ALT2

            return k

        elif (x['Category'] == 'Re-work_Order'):
            
            target_Rework = 8 * x['WD_'+month]

            k = (x['Count']) / target_Rework

            return k 
        
        


    

   
    KPI_PTRSH = PreTRSH_KPI.copy()
    
    
        # Note: First we have to group dataframe by names to get the total count and then merge the adjust_kpi dataframe
# to load Working_day variable for kpi calculation


    



    KPI_PTRSH = KPI_PTRSH.merge(adjusted_kpi, on='Names')


    KPI_PTRSH['KPI_Pre-TRSH'] = KPI_PTRSH.apply(KPI_preTRSH, axis=1)
    
    
    KPI_PTRSH = (KPI_PTRSH.groupby("Names")
               .sum()
               .reset_index()
               .loc[:, ["Names", "KPI_Pre-TRSH"]]

              )


    KPI_PTRSH['KPI_Pre-TRSH'] = np.round(KPI_PTRSH['KPI_Pre-TRSH'], 2)
    
    average = np.round(KPI_PTRSH['KPI_Pre-TRSH'].sum()/ len(PRETRSH_NAME),2)


    g = gg.ggplot(KPI_PTRSH, gg.aes(x='Names', y='KPI_Pre-TRSH',fill="Names", label='KPI_Pre-TRSH')) + \
        gg.geom_col(stat="identity") + \
        gg.theme_classic() + \
        gg.geom_text(position=gg.position_stack(vjust=0.5), size = 8) + \
        gg.geom_hline(yintercept = average,
                          color='#b15928',    
                          size=1,            
                          linetype="dashed") + \
        gg.annotate("text", label = "Avg: " + str(average), color="#de2d26", size = 10, x = 1.0, y =0.8) + \
        gg.labs(title= 'Pre-Troubleshooters KPI Report' + ' (' + str(month) + ')  \n\n No of Days: ' + str(DAYS) ,
                x=' ',
                y="KPI Index"
               ) + \
        gg.theme(axis_text_x=gg.element_blank())



    gg.ggsave(filename='Pre-Trouble_shooters_KPI_'+ str(month) + '.png', plot = g, path = figpath)
    
    

    return preTRSH_report_data,pre_TRSH, pre, g


preTRSH_report_data, pre_TRSH, pre_TRSH_fig_1, pre_TRSH_fig_2 = pre_TRSH_Performance_KPI(merged, figpath,PRETRSH_NAME, DAYS, adjusted_kpi)  # RDpt_lg is used to generate repair department line graph


print(pre_TRSH_fig_1)
print(pre_TRSH_fig_2)

pre_TRSH
```

## MVI Performance and KPI

```python
def MVI_Performance_KPI(merged, figpath, MVI_NAME, DAYS, adjusted_kpi):
    
    
    tmp = merged.copy()
    
    """ 
    Here we are removing all cases where STATUS is equal to prescrap as this information should not be
    included in the calculation of KPI
    
    """

    
    def prescrap(x):
        

        if (x['STATUS'] == 'Pre-Scrap'):
            return np.nan

        else: 
            return x["FAIL_Category"]


    tmp['RemovePrescrap']   =  tmp.apply(prescrap, axis=1)  


    cat_condition=["MVI", "SMA_Repair", "ICO_TRSH", 'Re-work_Order']
    
    

    mvi = (tmp.loc[(tmp['PRE-TRSH_Person'].isin(MVI_NAME)) & (tmp['RemovePrescrap'].isin(cat_condition))]
           .drop_duplicates(['RBU_SN',"RemovePrescrap",'PRE-TRSH_Person'], keep='last')
           .dropna(subset=['ProductNo'])
           .groupby(['PRE-TRSH_Person',"RemovePrescrap"])
           .count()
           .reset_index()
           .loc[:, ['PRE-TRSH_Person','RemovePrescrap', 'DATE']]
           .rename(columns={'PRE-TRSH_Person': 'Names',
                                   'RemovePrescrap':"Category",
                                   'DATE': "Count"})

             )



   
    mvi_plot = gg.ggplot(mvi, gg.aes(x='Names', y='Count',fill="Category", label='Count')) + \
                gg.geom_col(stat="identity", size=3) + \
                gg.theme_bw() + \
                gg.coord_flip() + \
                gg.geom_text(gg.aes(y = "Count",label = "Count"),  
                         size=8, nudge_x=0, nudge_y=20, show_legend=True) + \
                gg.labs(title= 'MVI Report' + ' (' + str(month) + ')  \n\n No of Days: ' + str(DAYS)) + \
                gg.theme(figure_size=(10, 6))



    mvi_plot= mvi_plot + gg.facet_wrap('~ Category', nrow=2)




    gg.ggsave(filename='MVI_performance_modified_'+ str(month) + '.png', plot = mvi_plot, path = figpath)
    
    
    
    
    #min_threshold_board_mvi=170
    
    #target_board_mvi=40 * DAYS  # 8 min /board




    def mvi_kpi(x):

        if (x['Category'] == 'MVI') | (x['Category'] == 'SMA_Repair') | (x['Category'] == 'ICO_TRSH') | (x['Category'] == 'MBFT') | (x['Category'] == 'Re-work_Order'):
            return "board_mvi"



    def KPI_MVI(x):

        if (x['Category'] == "board_mvi"):
            
            target_board_mvi = 41 * x['WD_'+month]

            k = (x['Count'])/ (target_board_mvi)

            return np.round(k, 2)


    
        
    KPI_mvi = mvi.copy()
    
    
    

    KPI_mvi['Category'] = KPI_mvi.apply(mvi_kpi, axis=1)
    
    # Note: First we have to group dataframe by names to get the total count and then merge the adjust_kpi dataframe
# to load Working_day variable for kpi calculation


    KPI_mvi = (KPI_mvi.groupby(["Names", "Category"])
           .sum()
           .reset_index()

          )




    KPI_mvi = KPI_mvi.merge(adjusted_kpi, on='Names')





    KPI_mvi['KPI_MVI'] = KPI_mvi.apply(KPI_MVI, axis=1)



    

    
    average = np.round(KPI_mvi['KPI_MVI'].sum()/ len(MVI_NAME),2)






    mvi_kpi = gg.ggplot(KPI_mvi, gg.aes(x='Names', y='KPI_MVI',fill="Names", label='KPI_MVI')) + \
        gg.geom_col(stat="identity") + \
        gg.theme_classic() + \
        gg.geom_text(position=gg.position_stack(vjust=0.5), size = 8) + \
        gg.geom_hline(yintercept = average,
                          color='#b15928',    
                          size=1,            
                          linetype="dashed") + \
        gg.annotate("text", label = "Avg: " + str(average), color="#de2d26", size = 10, x = 1.0, y =1.0) + \
        gg.labs(title= 'KPI_MVI' + ' (' + str(subdir) + ')  \n\n No of Days: ' + str(DAYS),
                x=' ',
                y="KPI Index"
               ) + \
        gg.theme(
        axis_line=gg.element_line(size=1, colour="black"),
        panel_grid_major=gg.element_line(colour="#d3d3d3"),
        panel_grid_minor=gg.element_blank(),
        panel_border=gg.element_blank(),
        panel_background=gg.element_blank(),
        plot_title=gg.element_text(size=13, family="Arial"),
        text=gg.element_text(family="Arial", size=10),
        axis_text_x=gg.element_text(colour="black", size=10),
        axis_text_y=gg.element_text(colour="black", size=10)
        
    )




    gg.ggsave(filename='KPI_MVI_'+ str(subdir) + '.png', plot = mvi_kpi, path = figpath)
    
    return mvi, mvi_plot, mvi_kpi





mvi_data, mvi_fig1, mvi_fig2 = MVI_Performance_KPI(merged, figpath, MVI_NAME, DAYS,adjusted_kpi)

print(mvi_fig1)
print(mvi_fig2)
```

## Soldering Performance and KPI

```python
def soldering_Performance_KPI(merged, figpath, REPAIR_NAME, DAYS, adjusted_kpi):
    
      
    tmp = merged.copy()

    """ 
    Here we are removing all cases where STATUS is equal to prescrap as this information should not be
    included in the calculation of KPI

    """


    def prescrap(x):


        if (x['STATUS'] == 'Pre-Scrap'):
            return np.nan

        if (x["FAIL_Category"] == 'DIG') | (x["FAIL_Category"] == 'RF') | (x["FAIL_Category"] == 'GU') | (x["FAIL_Category"] == 'Burn-In_Test') | (x["FAIL_Category"] == 'External_Claim') | (x["FAIL_Category"] == 'VRT'):
            return 'Radio'

        else: 
            return x["FAIL_Category"]


    tmp['RemovePrescrap']   =  tmp.apply(prescrap, axis=1)  


    cat_condition=["SMA_Repair", "MBFT", "ICO_TRSH", 'Re-work_Order', 'Radio']




    repair = (tmp.loc[(tmp['REPAIR_Person'].isin(REPAIR_NAME)) & (tmp['RemovePrescrap'].isin(cat_condition))]
           .drop_duplicates(['RBU_SN',"RemovePrescrap",'REPAIR_Person', 'REPAIR_NOTES'], keep='last')
           #.dropna(subset=['ProductNo']) 
           .groupby(['REPAIR_Person',"RemovePrescrap"])
           .count()
           .reset_index()
           .loc[:, ['REPAIR_Person','RemovePrescrap', 'DATE']]
           .rename(columns={'REPAIR_Person': 'Names',
                                   'RemovePrescrap':"Category",
                                   'DATE': "Count"})

             )


    soldering_plot = gg.ggplot(repair, gg.aes(x='Names', y='Count', fill="Category", label='Count')) + \
            gg.geom_bar(stat="identity") + \
            gg.theme_bw() + \
            gg.coord_flip() + \
            gg.geom_text(gg.aes(y = "Count",label = "Count"),  
                         size=8, nudge_x=0, nudge_y=2, show_legend=True) + \
            gg.labs(title= 'Soldering Report' + ' (' + str(month) + ')  \n\n No of Days: ' + str(DAYS)) + \
            gg.theme(figure_size=(10, 6))


    soldering_plot = soldering_plot + gg.facet_wrap('~ Category', nrow=2)



    gg.ggsave(filename='soldering_performance' + str(month) + '.png', plot = soldering_plot, path = figpath)






    # ## Calculating Soldering KPI



    soldering_categories=['Passive_Comp', '1_IC', '2_IC', 
                          'Connector', 'Pre-Scrap','3_IC']


    target_connector = 15 * DAYS     ## 28 min per connector / 1IC
    target_2_3IC = 7 * DAYS         ## 60 min per 2_IC or 3_IC
    target_1IC_Passive = 24 * DAYS   ## 18 min per manual soldering/hot air station for IC's or Passive Comp.


    def kpi_soldering(x):

        if x['Category'] == 'Connector':

            target_connector = 15 * x['WD_'+month]

            k = (x['Count']) / target_connector

            return k

        elif (x['Category'] == '2_IC') | (x['Category'] == '3_IC'):

            target_2_3IC = 7 * x['WD_'+month]

            k = (x['Count']) / target_2_3IC

            return k


        elif (x['Category'] == '1_IC') | (x['Category'] == 'Passive_Comp'):

            target_1IC_Passive = 24 * x['WD_'+month]

            k = (x['Count']) / target_1IC_Passive

            return k




    soldering_KPI = (tmp.loc[(tmp['REPAIR_Person'].isin(REPAIR_NAME)) & (tmp['RemovePrescrap'].isin(cat_condition))]
                     .drop_duplicates(['RBU_SN',"RemovePrescrap",'REPAIR_Person', 'REPAIR_NOTES'], keep='last')
                     .groupby(['REPAIR_Person', 'REPAIR_NOTES'])
                     .count()
                     .reset_index()
                     .loc[:, ['REPAIR_Person', 'REPAIR_NOTES', 'DATE']]
                     .rename(columns={'REPAIR_Person': 'Names',
                                     'REPAIR_NOTES': 'Category',
                                     'DATE': 'Count'})

                    )




    soldering_KPI = soldering_KPI.merge(adjusted_kpi, on='Names')



    soldering_KPI['Soldering KPI'] = soldering_KPI.apply(kpi_soldering, axis=1)



    soldering_KPI = (soldering_KPI.groupby('Names')
                    .sum()
                    .reset_index()
                    )



    soldering_KPI['Soldering KPI'] = np.round(soldering_KPI['Soldering KPI'], 2)

    average = np.round(soldering_KPI['Soldering KPI'].sum()/ len(REPAIR_NAME), 2)

    soldering_kpi = gg.ggplot(soldering_KPI, gg.aes(x='Names', y='Soldering KPI', fill='Names', label='Soldering KPI')) + \
        gg.geom_bar(stat="identity") + \
        gg.theme_classic() + \
        gg.geom_text(position=gg.position_stack(vjust=0.5), size = 8) + \
        gg.geom_hline(yintercept = average,
                          color='#b15928',    
                          size=1,            
                          linetype="dashed") + \
        gg.annotate("text", label = "Avg: " + str(average), color="#de2d26", size = 10, x = 1.0, y =0.8) + \
        gg.labs(title= 'Soldering KPI'+ ' (' + str(month) +')  \n\n No of Days: ' + str(DAYS))




    
    return repair, soldering_plot, soldering_kpi


repair_data, repair_fig1, repair_fig2 = soldering_Performance_KPI(merged, figpath, REPAIR_NAME, DAYS, adjusted_kpi)

print(repair_fig1)
print(repair_fig2)
```

```python
def radio_report(merged, TRSH_report_data, preTRSH_report_data, figpath):
    
    radios=["RF", "DIG"]

    tmp = (TRSH_report_data
        .rename(columns={"RemovePrescrap": "Category"})
        .query("Category in @ radios")
            )

    cat_preTRSH = ['Mech', 'ALT2', "Re-work_Order"]

    tmp2 = (preTRSH_report_data
        .query("Category in @ cat_preTRSH")
            )


      ## Taking remaining categories from merged file

    cat_3 = ['Burn-In_Test', 'External_Claim']

    tmp3 =  (merged
             .query('FAIL_Category in @ cat_3 ')
             .drop_duplicates(subset= ["RBU_SN", 'FAIL_Category'])
            )

    radio = pd.concat([tmp, tmp2, tmp3])

    radio_HC = radio.copy()
    ## we already removed duplicates from tmp, tmp2 and tmp3 therefore we have disabled drop_duplicates and is verified

    radio = (radio
                 #.drop_duplicates(subset=["RBU_SN", "FAIL_Category"], keep='last')  ## drop duplicate units and keeping the last row with unique unit
                 .groupby('DATE')
                 .count()
                 .reset_index()
                .loc[:, ['DATE','RBU_SN']]
                .rename(columns={
                                 'RBU_SN': "Count_Radio",
                                })

                )

    # Counting total no of attempts

    cat_total = ["RF", "DIG", 'Mech', 'ALT2', "Re-work_Order", 'Burn-In_Test', 'External_Claim']


    def new_category(x):

        if (x['FAIL_Category'] == "Mech") & (x['STATUS'] == "ALT2"):
                return "ALT2"

        elif (x['FAIL_Category'] == "Mech") & (x['STATUS'] != "ALT2"):
                return "Mech"

        else:
            return x['FAIL_Category']

    total_attempts = merged.copy()   
    total_attempts['Category'] = total_attempts.apply(new_category,axis=1)

    total_attempts = (total_attempts.query("Category in @ cat_total")
                       .groupby(['DATE','RBU_SN'])
                       .count()
                       .reset_index()
                       .groupby('DATE')
                       .sum()
                       .reset_index()
                       .loc[:, ['DATE','FAIL_Category']]
                       .rename(columns={'FAIL_Category': "Count_Total"})

                      )





    radio['DATE'] = radio['DATE'].astype('str')


    dates_mod = (pd.DataFrame(merged['DATE']
                              .astype(str)
                              .unique()
                              .tolist()
                              , columns=['DATE'])
                )


    radio= (radio.merge(dates_mod, how='outer', on = 'DATE').fillna(value=0))

    radio = radio.sort_values(by='DATE',ascending=True).reset_index(drop=True)


    average_radio = np.round(np.average(radio.query("Count_Radio != 0.0")['Count_Radio']),2)

    average_total_attempts = np.round(np.average(total_attempts.query("Count_Total != 0.0")['Count_Total']),2)

    total_attempts['DATE'] = total_attempts['DATE'].astype('str')
    total_attempts = total_attempts.merge(dates_mod, how='outer', on = 'DATE')


    total_attempts = total_attempts.sort_values(by='DATE',ascending=True).reset_index(drop=True)


    total_attempts = total_attempts.replace(np.nan, 0.0)




    from matplotlib.ticker import MaxNLocator
    plt.rcParams.update({'font.size': 12})



    fig, ax = plt.subplots(figsize=(20, 10))


    x = np.array(radio['DATE'])
    y = np.array(radio['Count_Radio'])

    y1 = np.array(total_attempts['Count_Total'])

    values = list(radio['Count_Radio'])
    values1 = list(total_attempts['Count_Total'])

    plt.xlabel('DATE')
    plt.ylabel('RADIO COUNT')


    plt.title('Repair Department Report'+ ' (' + str(month) + ')', fontsize=25)

    ax.plot(x, y, color='#2ca25f', linestyle='solid', label='PASS RADIO')

    ax.axhline(average_radio, color="#a1d99b", linestyle='dashed', linewidth=3)

    ax.plot(x, y1, color='#c51b8a', linestyle='solid', label='TOTAL ATTEMPTS')

    ax.axhline(average_total_attempts, color="#f768a1", linestyle='dashed', linewidth=3)

    dates_label = merged.DATE.astype('str').unique().tolist()

    ax.set_xticklabels(dates_label, rotation = 45, ha="right", position=(0,-0.01))



    for i, v in enumerate(values):
        plt.annotate(str(v), xy=(i,v), xytext=(-7,7), textcoords='offset points')

    for i, v in enumerate(values1):
        plt.annotate(str(v), xy=(i,v), xytext=(-7,7), textcoords='offset points')

    plt.ylim(0, 120)

    x_pos = -2.5
    y_pos = average_radio+3

    plt.text(x_pos, y_pos, "Avg: "+str(average_radio), color='green')

    plt.legend(['PASS RADIO', 'Avg. PASS RADIO','TOTAL ATTEMPTS', 'Avg. TOTAL ATTEMPTS'], fontsize=12)

    plt.grid()
    plt.show()

    fig.savefig(os.path.join(figpath, "Repair_department_weekly_report_Radio_" + str(month) +'.png'))


    
    return ax, radio_HC

ax, radio_HC = radio_report(merged, TRSH_report_data, preTRSH_report_data, figpath)
```
